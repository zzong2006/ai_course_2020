## 질문) `loc` 과 `column` indexing을 통한 performance 비교

indexing 결과가 `pd.DataFrame`이면, `pd.Series` 보다 오래 걸린다.
이 외에는 동일한 performance

````
df['a'][[0, 3, 7, 10, 12, 19]]
````

````
df.loc[[0, 3, 7, 10, 12, 19], 'a']  # Series 객체 반환
````

````
df.loc[[0, 3, 7, 10, 12, 19], ['a']] # DataFrame 객체 반환
````

## Matplotlib 를 사용하는 방법
1. 객체 지향 방식: 명시적으로 figures 와 axes 객체를 생성하고, 이들의 메소드를 호출함
2. `pyplot` 이라는 함수를 이용하여, plotting (`pyplot`은 자동으로 figures와 axes를 관리해줌)

둘 다 좋은 방법임. 하지만, 둘 중 하나를 사용하기로 결심했다면, 둘의 방법을 혼용하지 말것.

일반적으로, colab 과 같은 interactive plotting 방식에서는 `pyplot`을 이용하는 것을 추천     
그러나, 대형 프로젝트를 다룰 경우, non-interactive plotting 한 상황에서는 객체 지향 방식을 추천


### 1. 객체 지향 방식

````
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()               # figure 하나와 axes 하나 생성
ax.plot(x, x, label ='linear')        # axes 에 데이터 첫번째 plot
ax.plot(x, x ** 2, label ='quadratic')    # axes 에 데이터 두번째 plot
ax.plot(x, x ** 3, label ='cubic')    # axes 에 데이터 세번째 plot
ax.set_xlabel('x label')              # axes 에 x-label 을 추가
ax.set_ylabel('y label')              # axes 에 y-label 을 추가
ax.set_title("Simple Plot")           # axes 에 제목을 추가
ax.legend()                           # legend(범례) 추가
````

### 2. `pyplot`-style


````
x = np.linspace(0, 2, 100)

plt.plot(x, x, label ='linear')        # 데이터를 암묵적 (implicit) axes 에 첫번째 plot
plt.plot(x, x ** 2, label ='quadratic')    # axes 에 데이터 두번째 plot
plt.plot(x, x ** 3, label ='cubic')    # axes 에 데이터 세번째 plot
plt.xlabel('x label')              # axes 에 x-label 을 추가
plt.ylabel('y label')              # axes 에 y-label 을 추가
plt.title("Simple Plot")           # axes 에 제목을 추가
plt.legend()                           # legend(범례) 추가
````

## `Pyplot`
`matplotlib.pyplot`는 함수의 모음(a collection of functions)이다. 각 함수는  하나의 figure에 변화를 준다.    
(예를 들어, figure를 생성하거나, figure에 plotting할 공간(axes)을 설정하거나, 그 공간에 선을 긋거나, label을 이용해 plot 된 내용을 꾸미거나 등) 

### `plot` 함수
`pyplot`을 통해 다음과 같이 빠르게 시각화가 가능

````
plt.plot([1, 2, 3, 4])    # 하나의 array (또는 list) 를 건네준 경우, 그 값은 y 축의 값으로 인식하여 자동으로 x 에 해당하는 값을 생성
# x 축 값은 [0, 1, 2, 3] (python 은 0부터 시작하므로)
plt.ylabel('some numbers')
plt.show()
````

### 여러개의 figures와 axes로 작업하기

`pyplot`은 plot을 위한 figure 또는 axes에 관한 메소드를 호출해주기 전 까지는 figure와 현재 axes 정보를 저장하여, 지속적으로 그 공간에 plotting 수행

아래 함수 `f(t)`는 다음을 수행
$$
f(t) = e^{t} \cdot cos(2 \pi t)
$$

````
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()        # figure 를 수행한 순간 subplot(111)이 수행됨
plt.subplot(211)    # 2 by 1 gird 중에 첫번째 grid로 axes 변환, This method call is equal to `subplot(2, 1, 1)`
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)    # 2 by 1 grid 중에 두번째 grid로 axes 변환
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
````

````
import matplotlib.pyplot as plt
plt.figure(1)                # 첫번째 figure
plt.subplot(211)             # 첫번째 figure의 첫번째 subplot 
plt.plot([1, 2, 3])
plt.title('First Figure')
plt.subplot(212)             # 첫번째 figure의 두번째 subplot 
plt.plot([4, 5, 6])


plt.figure(2)                # 두번째 figure
plt.plot([4, 5, 6])          # subplot(111)이 자동적(by default)으로 수행됨 
plt.title('Second Figure')

plt.show()
````

### Logarithmic axes

특정 축 (axis)의 간격을 linear axis scales 이 아니라, logarithmic scales로 바꿀 수 잇음

````
# seed 설정 for reproducibility
np.random.seed(19680801)

# 0과 1 사이의 값에 대한 정규 분포 샘플링: loc (평균), scale (표준 편차)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]    
y.sort()            
x = np.arange(len(y))

plt.figure()

# linear
plt.subplot(121)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(122)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# subplot 간 간격을 조절함 (log scale로 인한 표기 때문)
"""
  left, bottom, right, top
  서브플롯 4면의 위치 조정

  hspace, wspace
  서브플롯 간의 간격을 위해 각각 figure의 너비와 높이에 대한 비율 조절
"""
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
````

#### `boxplot` 을 이용한 분포 관측

![images](https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=http%3A%2F%2Fcfile1.uf.tistory.com%2Fimage%2F99D3C43359C29437228FD1)

박스 플롯은 수치적 자료를 표현하는 그래프이다. 이 그래프는 자료에서 얻은 다섯 수치 요약(five number summary)을 가지고 그린다.

다섯 수치 요약은 아래와 같다.

1. 최솟값 : 제 1사분위에서 1.5 IQR1을 뺀 위치이다.
2. 제 1사분위(Q1) : 25%의 위치를 의미한다.
3. 제 2사분위(Q2) : 50%의 위치로 중앙값(median)을 의미한다.
4. 제 3사분위(Q3) : 75%의 위치를 의미한다.
5. 최댓값 : 제 3사분위에서 1.5 IQR을 더한 위치이다.

최솟값과 최댓값을 넘어가는 위치에 있는 값을 이상치(Outlier)라고 부른다.

````
# catplot 에서 kind="box" 사용
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
````

````
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips)
````
